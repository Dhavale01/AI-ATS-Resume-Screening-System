```python
import pandas as pd
import ast
```

```python
ranked_candidates = pd.read_csv(
    r"D:\AI_ATS_Resume_Screening_System\resume\test\csv\candidate_ranking.csv"
)

ranked_candidates.head()
```

```python
job_skills = {

    "python",
    "sql",
    "machine learning",
    "pandas",
    "numpy",
    "statistics",
    "data analysis",
    "power bi"

}
```

```python
def convert_to_list(skill_string):

    try:
        return ast.literal_eval(skill_string)

    except:
        return []

ranked_candidates["skills_found"] = (
    ranked_candidates["skills_found"]
    .apply(convert_to_list)
)
```

```python
def analyze_skill_gap(candidate_skills):

    candidate_skills = set(candidate_skills)

    matched_skills = (
        job_skills.intersection(candidate_skills)
    )

    missing_skills = (
        job_skills.difference(candidate_skills)
    )

    return (
        list(matched_skills),
        list(missing_skills)
    )
```

```python
results = ranked_candidates["skills_found"].apply(
    analyze_skill_gap
)

ranked_candidates["matched_skills"] = (
    results.apply(lambda x: x[0])
)

ranked_candidates["missing_skills"] = (
    results.apply(lambda x: x[1])
)
```

```python
ranked_candidates["missing_skill_count"] = (
    ranked_candidates["missing_skills"]
    .apply(len)
)
```

```python
ranked_candidates[
    [
        "resume_name",
        "match_percentage",
        "matched_skills",
        "missing_skills"
    ]
].head()
```

```python
def generate_recommendation(missing_skills):

    if len(missing_skills) == 0:

        return (
            "Excellent match for this role."
        )

    return (
        "Recommended to learn: "
        + ", ".join(missing_skills)
    )
```

```python
ranked_candidates["recommendation"] = (
    ranked_candidates["missing_skills"]
    .apply(generate_recommendation)
)
```

```python
top_candidate = ranked_candidates.iloc[0]

print("Candidate:")
print(top_candidate["resume_name"])

print("\nMatch Percentage:")
print(top_candidate["match_percentage"])

print("\nMatched Skills:")
print(top_candidate["matched_skills"])

print("\nMissing Skills:")
print(top_candidate["missing_skills"])

print("\nRecommendation:")
print(top_candidate["recommendation"])
```

```python
ranked_candidates.to_csv(
    r"D:\AI_ATS_Resume_Screening_System\resume\test\csv\final_ats_results.csv",
    index=False
)

print("Final ATS results saved!")
```
