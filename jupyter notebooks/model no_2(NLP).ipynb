```python
import pandas as pd
import numpy as np

import nltk
import spacy
import re
import string

from nltk.corpus import stopwords
```

```python
resume_df = pd.read_csv(
    r"D:\AI_ATS_Resume_Screening_System\resume\test\csv\resume_data.csv"
)

resume_df.head()
```

```python
nlp = spacy.load("en_core_web_sm")

stop_words = set(stopwords.words('english'))

print("NLP resources loaded successfully!")
```

```python
print(resume_df["resume_text"][0][:3000])
```

```python
def clean_resume_text(text):

    # Convert to string
    text = str(text)

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove phone numbers
    text = re.sub(r'\d{10,}', '', text)

    # Remove punctuation
    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    return text.strip()
```

```python
resume_df["clean_text"] = resume_df["resume_text"].apply(
    clean_resume_text
)

resume_df.head()
```

```python
print("BEFORE:\n")
print(resume_df["resume_text"][0][:1000])

print("\n" + "="*100 + "\n")

print("AFTER:\n")
print(resume_df["clean_text"][0][:1000])
```

```python
def remove_stopwords(text):

    words = text.split()

    filtered_words = [
        word
        for word in words
        if word not in stop_words
    ]

    return " ".join(filtered_words)
```

```python
resume_df["clean_text"] = resume_df["clean_text"].apply(
    remove_stopwords
)
```

```python
def lemmatize_text(text):

    doc = nlp(text)

    lemmas = []

    for token in doc:

        if not token.is_space:

            lemmas.append(token.lemma_)

    return " ".join(lemmas)
```

```python
resume_df["clean_text"] = resume_df["clean_text"].apply(
    lemmatize_text
)
```

```python
print("ORIGINAL:\n")
print(resume_df["resume_text"][0][:1000])

print("\n" + "="*100 + "\n")

print("CLEANED:\n")
print(resume_df["clean_text"][0][:1000])
```

```python
resume_df.to_csv(
    r"D:\AI_ATS_Resume_Screening_System\resume\test\csv\clean_resume_data.csv",
    index=False
)

print("Clean dataset saved successfully!")
```
